```python
# hf demo for comparison --> https://huggingface.co/spaces/abidlabs/music-separation
# where this inference function is taken from

import os
import numpy as np
from pydub import AudioSegment
from scipy.io.wavfile import read, write


def inference(input_filepath):
    # convert mp3 to wav
    temp_wav = "data/output/temp.wav"
    os.system(f"rm {temp_wav}")
    os.system(f"ffmpeg -i {input_filepath} -f wav {temp_wav}")
    
    # read in temp wav file
    sample_rate, data = read(temp_wav)
    
    # write to output
    os.system(f"python3 -m demucs.separate -n htdemucs --two-stems=vocals {temp_wav} -o out")

    os.system(f"rm {temp_wav}")

    return "./data/output/vocals.wav","./data/output/no_vocals.wav"
```

```python
input_filepath = "./data/input/boom_short.mp3"
data = inference(input_filepath)
```
